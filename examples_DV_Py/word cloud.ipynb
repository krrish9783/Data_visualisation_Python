```python
from wordcloud import WordCloud
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('/home/krishna/Documents/sales.csv')

# Extract relevant columns
countries = data['Country']
profits = data['Total Profit']

# Create a dictionary of country:profit pairs
country_profit_dict = dict(zip(countries, profits))

# Filter countries with higher profits
high_profits = {country: profit for country, profit in country_profit_dict.items() if profit >= 5000}  # Adjust the threshold as needed

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400).generate_from_frequencies(high_profits)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Countries with Higher Profits')
plt.show()
```
